```python
import random
from collections import Counter
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
Age = [random.randint(18,45) for _ in range(400)]
Monthly_Income = [random.randint(25000, 100000) for _ in range(400)]
gen = ['Male', 'Female']
Gender= [random.choice(gen) for _ in range(400)]
loc = ['Pune', 'Mumbai','Banglore', 'Delhi']
Location= [random.choice(loc) for _ in range(400)]
```

```python
df = {}
df["Gender"] = Gender
df["Age"] = Age
df['Monthly_Income'] = Monthly_Income
df['Location'] = Location
```

```python
dataFrame = pd.DataFrame(df)
```

```python
# Convert the data to lists
ages = dataFrame["Age"]
incomes = dataFrame["Monthly_Income"]

# Calculate the summary statistics for age
n = len(ages)
age_mean = sum(ages) / n
age_median = sorted(ages)[n // 2]
age_min = min(ages)
age_max = max(ages)
age_count=ages.count()
age_percentile25 = np.percentile(ages, 25)
age_percentile50 = np.percentile(ages, 50)
age_percentile75 = np.percentile(ages, 75)
age_diff = [(age - age_mean) ** 2 for age in ages]
age_var = sum(age_diff) / (n - 1)
age_std = age_var ** 0.5

# Calculate the summary statistics for income
n = len(incomes)
income_mean = sum(incomes) / n
income_median = sorted(incomes)[n // 2]
income_min = min(incomes)
income_max = max(incomes)
income_count=incomes.count()
income_percentile25 = np.percentile(incomes, 25)
income_percentile50 = np.percentile(incomes, 50)
income_percentile75 = np.percentile(incomes, 75)
income_diff = [(income - income_mean) ** 2 for income in incomes]
income_var = sum(income_diff) / (n - 1)
income_std = income_var ** 0.5

# Print the summary statistics
print("Age Summary Statistics:")
print("Mean: ",age_mean)
print("Median: ",age_median)
print("Minimum: ",age_min)
print("Maximum: ",age_max)
print("Count: ",age_count)
print("25%: ",age_percentile25)
print("50%: ",age_percentile50)
print("75%: ",age_percentile75)
print(f"Variance: {age_var:.2f}")
print(f"Standard Deviation: {age_std:.2f}")
print()
print("Monthly Income Summary Statistics:")
print("Mean: ",income_mean)
print("Median: ",income_median)
print("Minimum: ",income_min)
print("Maximum: ",income_max)
print("Count: ",income_count)
print("25%: ",income_percentile25)
print("50%: ",income_percentile50)
print("75%: ",income_percentile75)
print(f"Variance: {income_var:.2f}")
print(f"Standard Deviation: {income_std:.2f}")


# Create a dictionary of the summary statistics
summary_dict = {
    "Statistic": ["Mean", "Median", "Minimum", "Maximum","Count","25%","50%","75%","Variance","Standard Deviation"],
    "Age": [age_mean, age_median, age_min, age_max,age_count,age_percentile25,age_percentile50,age_percentile75,age_var,age_std],
    "Monthly Income": [income_mean, income_median, income_min, income_max,income_count,income_percentile25,income_percentile50,income_percentile75,income_var,income_std]
}

# Convert the dictionary to a pandas DataFrame
summary_df = pd.DataFrame(summary_dict)

# Set the "Statistic" column as the index
summary_df.set_index("Statistic", inplace=True)

# Print the summary statistics DataFrame
summary_df
```
